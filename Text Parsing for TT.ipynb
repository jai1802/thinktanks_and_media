```python
think_tank_dic = {'American Enterprise Institute': ['American Enterprise Institute', 'A.E.I', 'AEI']}
```

```python
saying_list = ['said', 'say', 'explained', 'explains', 'according', 'report', 'replied', 'replies', 'study', 'told', 'tells']
```

```python
def get_sentence(text, tt):
    # split into sentences with nltk
    # check each sentence for think tank name and a.k.as along with saying verb in each sentence
    if not text or not isinstance(text, str):
        return None
    for sentence in tokenizer.tokenize(text):
        if any([i in sentence for i in think_tank_dic[tt]]) and any([said_word in sentence for said_word in saying_list]):
            return sentence
    return None
```

```python
title_arr, text_arr, date_arr, tt_arr, sentence_arr, pub_arr = [], [], [], [], [], []
for i in think_tank_citations.index:
    title, text, date = think_tank_citations['title'][i], think_tank_citations['article'][i], think_tank_citations['date'][i]
    pub = think_tank_citations['pub'][i]
    for tt in think_tank_dic:
        match = get_sentence(text, tt)
        if match:
            title_arr.append(title)
            date_arr.append(date)
            tt_arr.append(tt)
            sentence_arr.append(match)
            pub_arr.append(pub)
```

```python
df = pd.DataFrame({'title':title_arr, 'date':date_arr, 'thinktank':tt_arr, 
                   'sentence':sentence_arr, 'publication':pub_arr})
```
