```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import requests
```

# Name List to JSON

```python
df = pd.read_csv("./data/thinktank_data/thinktank_features.csv", index_col=0)
```

```python
tt_name_dic = {}
for i in df.index:
    tt_name = df['thinktank'][i]
    name_list = list(df[['name1', 'name2', 'name3']].iloc[i,:])
    tt_name_dic[tt_name] = name_list
```

```python
len(tt_name_dic)
```

```python
# writing to json file
```

```python
import json
with open('./data/thinktank_data/tt_names.json', 'w') as f:
    json.dump(tt_name_dic, f)
```
